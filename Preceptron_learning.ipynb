```python
# Perceptron to learn AND logic gate

# Step 1: Training data
inputs = [
    [0, 0],
    [0, 1],
    [1, 0],
    [1, 1]
]

labels = [0, 0, 0, 1]  # AND gate outputs

# Step 2: Initialize weights and bias
weights = [0.0, 0.0]
bias = 0.0
learning_rate = 0.1
epochs = 10  # Number of times we go through the data

# Step 3: Activation function
def step_function(x):
    return 1 if x > 0 else 0

# Step 4: Training loop
for epoch in range(epochs):
    print(f"\nEpoch {epoch+1}")
    for x, y in zip(inputs, labels):
        weighted_sum = x[0]*weights[0] + x[1]*weights[1] + bias
        prediction = step_function(weighted_sum)
        error = y - prediction

        # Update weights and bias
        weights[0] += learning_rate * error * x[0]
        weights[1] += learning_rate * error * x[1]
        bias += learning_rate * error

        print(f"Input: {x}, Prediction: {prediction}, Actual: {y}, Error: {error}")

print("\nFinal weights:", weights)
print("Final bias:", bias)
```
